# **Most Popular Starting point** 

```python
df1=df["START*"].dropna()
df2=pd.DataFrame(df1.value_counts())
df2.sort_values(["START*"],ascending=False)
df2=df2.reset_index()
df2=df2.rename(columns= {'index':'Start_Destination','START*':'Count'})
df2.loc[df2['Count'] == max(df2['Count'])]

```

# **MOst Popular Dropping point**

```python
df1=df["STOP*"].dropna()
df2=pd.DataFrame(df1.value_counts())
df2.sort_values(["STOP*"],ascending=False)
df2=df2.reset_index()
df2=df2.rename(columns={'index':'Destination ','STOP*':'Count'})
df2.loc[df2['Count']==min(df2["Count"])]
```

# **MOst Frequent taken route**

```python
df=df.dropna()
df1=pd.DataFrame(df.groupby(['START*','STOP*']).size())
df2=df1.rename(columns={0:'Count'})
df2=df2.sort_values(['STOP*'],ascending=False)
#df2.head(10)
df2.loc[df2['Count']==(df2['Count'])].head(10)
```
